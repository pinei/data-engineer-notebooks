## NumPy

- NumPy is a Python library used for working with arrays, linear algebra, fourier transform, and matrices. NumPy stands for Numerical Python and it is an open source project.
- A numpy array is similar to a list.
- The array object in NumPy is called **ndarray**, it provides a lot of supporting functions that make working with ndarray very easy.
- Arrays are very frequently used in data science, where speed and resources are very important.
- NumPy is usually imported under the np alias.
- It's usually fixed in size and each element is of the same type.

```python
import numpy as np 

print(np.__version__)
```

```python
a = np.array([0, 1.1, 2.4, 3.1, 4.0])

print(f'type: {type(a)}')
print(f'dtype: {a.dtype}')

print(f'size: {a.size}')
print(f'ndim: {a.ndim}')
print(f'shape: {a.shape}')

print(f'mean: {a.mean()}')
print(f'max: {a.max()}')
print(f'min: {a.min()}')
print(f'std: {a.std()}')
```

```python
# Filter
filter = [True, False, False, False, True]
selection = a[filter]
print(selection)

# Slice
slice = a[2:4]
print(slice)
```

```python
# Plotting functions

import time 
import sys
import numpy as np 
import matplotlib.pyplot as plt

%matplotlib inline  

def Plotvec1(u, z, v):
    # Full window axes
    ax = plt.axes()

    ax.arrow(0, 0, *u, head_width=0.05, color='r', head_length=0.1) 
    plt.text(*(u + 0.1), 'u')
    
    ax.arrow(0, 0, *v, head_width=0.05, color='b', head_length=0.1)
    plt.text(*(v + 0.1), 'v')
    
    ax.arrow(0, 0, *z, head_width=0.05, head_length=0.1)
    plt.text(*(z + 0.1), 'z')
    
    plt.ylim(-2, 3)
    plt.xlim(-2, 3)
```

```python
# Addition
u = np.array([1,0])
v = np.array([0,1])
z = np.add(u, v)

Plotvec1(u, z, v)
```

```python
# Subtraction
u = np.array([1,0])
v = np.array([0,1])
z = np.subtract(u, v)

Plotvec1(u, z, v)
```

```python
# Multiplication
x = np.array([1, 2])
y = np.array([2, 1])
z = np.multiply(x, y)

print(z)
Plotvec1(u, z, v)
```

```python
# Dot product
x = np.array([1, 2])
y = np.array([3, 2])
z = np.dot(x, y)

print(z)
```

### Linspace

A useful function for plotting mathematical functions is <code>linspace</code>.   Linspace returns evenly spaced numbers over a specified interval. 

 **numpy.linspace(start, stop, num = int value)**
 
start  :  start of interval range\
stop   :  end of interval range\
num    :  Number of samples to generate.

```python
np.linspace(-2, 2, num=5)
```

```python
x = np.linspace(0, 2*np.pi, num=100)
y = np.sin(x)
plt.plot(x, y)
```

```python
x = np.linspace(0, 2*np.pi, num=100)
y = np.cos(x)
plt.plot(x, y)
```

Multiplication of two arrays corresponds to an element-wise product or <em>Hadamard product</em>. The Hadamard product corresponds to multiplying each of the elements in the same position, i.e. multiplying elements contained in the same color boxes together.

```python
# Array Multiplication
diagonal = np.array([[1,0,0],[0,1,0],[0,0,1]])
b = np.multiply(a, diagonal)

print(b)
```

We can also perform matrix multiplication with Numpy arrays. Matrix multiplication is a little more complex.

```python
# Dot Multiplication
a = np.array([[0, 1, 1],[1, 0, 1]])
b = np.array([[1,1],[1,1],[-1,1]])

z = np.dot(a, b)
print(z)
```
